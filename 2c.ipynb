# Funções Seno e Cosseno
#### Essas funções são de suma importância, pois facilita o cálculo de áreas e ângulos ditos impossíveis.
#### Como aplicação, é muito comum encontrar em construções civis e em jogos.

# Seno

```python
import math
import numpy as np
from pylab import *
import matplotlib.pyplot as plt
```

```python
listasinx = np.arange(-1.0, 1.0, 0.01)
listasiny = sin(2*pi*listasinx)
```

```python
plt.plot(listasinx, listasiny)
```

# Cosseno

```python
listacosx = np.arange(-1.0, 1.0, 0.01)
listacosy = cos(2*pi*listacosx)
```

```python
plt.plot(listacosx, listacosy)
```
